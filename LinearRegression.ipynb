```python
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
X_Matrix = []
Y_Matrix = []
X_Label = []
Y_Label = []
PI = 3.141592654
```

```python
# 产生sinx的数据集
for i in np.arange(0, 2 * PI, PI / 20):
    x0 = i
    x1 = pow(i, 3)
    x2 = pow(i, 5)
    x3 = pow(i, 7)
    X_Label.append(x0)
    X_Matrix.append([x0, x1, x2, x3, 1])
    Y_Matrix.append([math.sin(x0)])
    Y_Label.append(math.sin(x0))
plt.title("Dataset points scatter plot")
plt.xlabel('X')
plt.ylabel('Sin(X)')
plt.scatter(X_Label, Y_Label, c='red')
plt.show()
```

```python
# 利用矩阵计算线性回归的梯度矩阵 公式w=(X^T*X)^-1 *X^T*y
# 构造矩阵X 和 转置矩阵XT
Matrix_X = np.matrix(X_Matrix)
Matrix_XT = np.transpose(Matrix_X)
# 计算矩阵(X^T*X)的逆 XI
Matrix_XTX = np.matmul(Matrix_XT, Matrix_X)
Matrix_XTXI = Matrix_XTX.I
# 计算权重矩阵Weights_pre,Weights_post
Weights_pre = np.matmul(Matrix_XTXI, Matrix_XT)
Weights_post = np.matrix(Y_Matrix)
# 计算权重矩阵Weights
Weights = np.matmul(Weights_pre, Weights_post)
# 计算Ordinary Least Squares方法的预测值
Y_Predict = np.matmul(Matrix_X, Weights)
Y_Predict = Y_Predict.tolist()
# 输出线性回归结果的表达式
print("Y = ", end="")
for i in range(len(Weights.tolist()) - 1):
    print("X" + str(i), ' * ', end="")
    print(Weights.tolist()[i][0], " + ", end="")
print(Weights.tolist()[len(Weights.tolist()) - 1][0])
print("MSE = ", mean_squared_error(Y_Label, Y_Predict))
print("MAE = ", mean_absolute_error(Y_Label, Y_Predict))
plt.title("Linear Regression Result")
plt.xlabel('X')
plt.ylabel('Sin(X)')
plt.scatter(X_Label, Y_Label, c='red')
plt.plot(X_Label, Y_Predict, c='blue')
plt.show()
```
